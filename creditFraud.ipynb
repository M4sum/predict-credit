```python
from math import sqrt
def distance(x,y):
    ans=0.00
    if(len(x)==len(y)):
        for i in range(len(x)):
            ans+=((x[i])-(y[i]))**2
    return (sqrt(ans))
```

```python
def knearestmodel(xtrain,ytrain,xtest,k):
    xtrain=np.array(xtrain)
    ytrain=np.array(ytrain)
    xtest=np.array(xtest)
    kl=0
    No_of_class=len(np.unique(ytrain))
    uniqueclass=np.unique(ytrain)
    
    ypred=[]
    
    for i in xtest:
        dist=[]
        for j in xtrain:
            dist.append(distance(i,j))
        dist=np.array(dist)
        ind=np.argsort(dist)
        voting=ind[:k]
        classvote=np.zeros(No_of_class)
        for f in range(len(voting)):
            ourvote=ytrain[voting[f]]
            for z in range(len(uniqueclass)):
                if(ourvote==uniqueclass[z]):
                    classvote[z]+=1
        prediction=uniqueclass[np.argmax(classvote)]
        ypred.append(prediction)
        print(kl)
        kl+=1
        del dist
        del ind
        del voting
        del classvote
        del ourvote
        
    return (np.array(ypred))
```
